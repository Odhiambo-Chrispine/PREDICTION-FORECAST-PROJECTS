# Importing Relevant Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

# Importing dataset

```python
data=pd.read_csv("horse.csv")
data
```

# Specifying the targeted variable and its unique attributes

```python
target=data["outcome"]
target.unique()
```

# Categorizing the animals into dummies

```python
data.isnull().sum()
```

```python
data2=data[['rectal_temp','pulse','respiratory_rate','packed_cell_volume','total_protein','lesion_1','lesion_2','lesion_3']]
```

# Categorizing the variables from string to float

```python
df=pd.get_dummies(data[["surgery","age","cp_data"]])
result=pd.concat([data,df],axis=1).reindex(data.index)
result
```

# Importing Libraries for Training and Splitting data

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

## Converting categorical attributes into numerical

```python
le=LabelEncoder()
label=le.fit_transform(data[["cp_data"]])
label
```

## Training and testing data

```python
x,y=data2.values,target.values
label_encoder=LabelEncoder()
y=label_encoder.fit_transform(y)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

# Imputing the missing values

```python
from sklearn.impute import SimpleImputer
import numpy as np
imp=SimpleImputer()
x_train=imp.fit_transform(x_train)
x_test=imp.fit_transform(x_test)
```

# Decision Tree Classification

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier=DecisionTreeClassifier()
classifier.fit(x_train,y_train)
```

## Predicting the test

```python
y_predict=classifier.predict(x_test)
```

## Checking the accuracy of the model

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_predict,y_test)
accuracy
```

# Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier2=RandomForestClassifier()
classifier2.fit(x_train,y_train)
y_prediction=classifier2.predict(x_test)
```

## Checking the accuracy of the model

```python
accuracy2=accuracy_score(y_prediction,y_test)
accuracy2
```
