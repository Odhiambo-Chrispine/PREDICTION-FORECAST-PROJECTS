# Importing Relevant Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# Importing Dataset

```python
data=pd.read_csv('driver-data.csv')
```

# Importing Relevant library for Kmeans

```python
from sklearn.cluster import KMeans
```

## Renaming the clustering

```python
Kmeans=KMeans(n_clusters=4)
data_analyze=data.drop('id',axis=1)
```

## Fitting the clusters into the data

```python
Kmeans.fit(data_analyze)
```

## Getting information from the cluster centers

```python
Kmeans.cluster_centers_
```

## Displaying the labels of tyhe kmeans and their lengths

```python
print(Kmeans.labels_)
print(len(Kmeans.labels_))
```

## Checking on the number of unique drivers are there in first and second cluster

```python
print(type(Kmeans.labels_))
unique,counts=np.unique(Kmeans.labels_,return_counts=True)
print(dict(zip(unique,counts)))
```

# Plotting the clusters

```python
data_analyze['cluster']=Kmeans.labels_
sns.set_style('whitegrid')
sns.lmplot('mean_dist_day','mean_over_speed_perc',data=data_analyze,
           hue='cluster',palette='coolwarm',size=6,aspect=1,fit_reg=False)
```
