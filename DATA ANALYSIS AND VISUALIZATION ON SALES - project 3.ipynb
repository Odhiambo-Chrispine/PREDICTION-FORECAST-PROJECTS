# PART TWO : VISUALIZATION

## Distribution of data - Distribution curve

```python
sns.distplot(data["Units Sold"],hist=False,color="green",kde=True)
```

## Distribution of data - Histogram

```python
sns.distplot(data["Units Sold"],hist=True,color="green",kde=False)
```

## Relation plot

```python
sns.relplot(x="Date",y="Units Sold",kind="line",data=data.head(60))
plt.xticks(rotation=75)
plt.show()
```

## Facet Grid

```python
sns.FacetGrid(data.head(50),col="Country").map(plt.hist,"Units Sold")
```

## Pair Grid

```python
sns.set(style="darkgrid")
sns.PairGrid(data.head(30)).map(plt.scatter,color="green")
plt.show()
```

## Cat plot

```python
sns.catplot(x="Date",y="Units Sold",kind="violin",data=data.head(20))
plt.xticks(rotation=75)
```

## Box plot

```python
sns.boxplot(x="Date",y="Units Sold",data=data.head(20))
plt.xticks(rotation=75)
```

## Mean Distribution of units sold

```python
mean=data["Units Sold"].mean()
sns.distplot(data["Units Sold"])
plt.axvline(mean,0,1,color="green")
```

## Line plot

```python
sns.lineplot(x="Country",y="Units Sold",hue="Segment",data=data.head(50),ci=None)
```

```python
sns.lineplot(x="Country",y="Units Sold",data=data.head(50),ci=False,style="Segment",marker=True)
plt.show()
```

## Pie Charts

```python
plt.pie(data["Units Sold"].head(5),explode=(0.0,0.4,0.2,0.0,0.0),autopct="%0.2f%%",labeldistance=1.2,radius=2.6, wedgeprops={"linewidth":3,"edgecolor":"red"})
plt.show()
```

## Bar Plot

```python
sns.barplot(x=data.Date.head(50),y=data["Units Sold"].head(50))
plt.xticks(rotation=75)
plt.show()
```

## Linear Regression

```python
sns.lmplot(x="Month Position",y="Units Sold",data=data.head(200),fit_reg=True,hue="Date")
```

## Pair Plot

```python
sns.pairplot(data.head(100),hue="Date",palette="Blues")
```

```python
sns.pairplot(data.head(50))
```

## Rug Plot

```python
sns.rugplot(data["Units Sold"])
```

## Heat Map

```python
sns.heatmap(data.corr(),cmap='Blues',annot=True)
```

## Count Plot

```python
sns.set()
sns.countplot('Country',data=data)
```

```python
sns.countplot('Country',hue='Units Sold',data=data)
```
